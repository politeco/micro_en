## Functions

It will be useful to create our own functions. Python functions are defined using the def statement:

```python
def myfunc(x):
    return x**2
```

```python
y = [myfunc(i) for i in range(10)]
y
```

Python has its own way of dealing with variable names. First variables defined (assigned) inside a function are local (i.e. they have no value outside the function or do not change the value of a variable outside). Take this example: 

```python
alpha = 2
```

And we construct: 

```python
def myfunc(x):
    return x**alpha
```

alpha is not defined inside the function. Python will look for it in variables outside the function. If it finds a variable that matches its type, it will use that value. We then call this variable a global variable.

```python
myfunc(5)
```

Suppose instead that we have:

```python
def myfunc(x):
    alpha = 3
    return x**alpha
```

```python
myfunc(2)
```

```python
alpha
```

What happened here? alpha takes value 2 on the outside, but inside it takes value 3. So it uses 3 to compute what myfunc returns, but does not change the global variable. 

In general we avoid global variables and use additional arguments: 


```python
def myfunc(x,alpha):
    return x**alpha
```

```python
myfunc(2,2), myfunc(2,3)
```

```python
alpha
```

A last important action is what to do with functions who take many arguments but we want to keep those other arguments fixed while we only vary one.  Some routines, in optimize for example, only need one variable functions. The partial function is very useful: 

```python
from functools import partial
```

```python
f = partial(myfunc,alpha=3.0)
```

```python
f(2)
```

```python
g = partial(myfunc,alpha=2.0)
```

```python
g(2)
```

Let's make sense of derivatives and optimization. 

### Drivative

Consider the function $f=(\alpha -\frac{\beta}{2}x)x$. 

```python
def f(x,alpha,beta):
    return (alpha - 0.5*beta*x) * x
```

```python
x = np.linspace(0,20,10)
alpha = 10
beta = 1.0
y = f(x,alpha,beta)
```

```python
plt.figure()
plt.plot(x,y)
plt.show()
```

We can observe how the function changes as we vary $\alpha$:

```python
alphas = np.linspace(5,15,10)
beta = 1.0
plt.figure()
for alpha in alphas: 
    plt.plot(x,f(x,alpha,beta),label=str(np.round(alpha,2)))
plt.legend(loc=2)
plt.title("Changer $\\alpha$")
plt.show()
```

Same thing or beta (note how we use latex to render greek letters in the plots): 

```python
betas = np.linspace(0.5,1.5,10)
alpha = 10.0
plt.figure()
for beta in betas: 
    plt.plot(x,f(x,alpha,beta),label=str(np.round(beta,2)))
plt.legend(loc=2)
plt.title("Changer $\\beta$")
plt.show()
```

Let's now program the derivative. The derivative is given by $$ f'(x) = \alpha - \beta x $$ 

```python
def df(x,alpha,beta):
    return alpha - beta*x
```

Graphically we have :

```python
alpha = 10
beta = 1.0
y = f(x,alpha,beta)
dy = df(x,alpha,beta)
plt.figure()
plt.plot(x,y,label='fonction')
plt.plot(x,dy,label='drive')
plt.axvline(x=10.0,linestyle=':')
plt.plot(x,np.zeros(len(x)),linestyle=':')
plt.show()
```

How we compute numerically the derivate? (suppose we are lazy, the derivative is very complicated or the function is unkown).  Yes, if we take small finite differences:

```python
def dfn(x,alpha,beta,eps=1e-6):
    return (f(x+eps,alpha,beta) - f(x-eps,alpha,beta))/(2*eps)
```

```python
alpha = 10
beta = 1.0
dy = df(x,alpha,beta)
dny = dfn(x,alpha,beta)
plt.figure()
plt.scatter(x,dny,label='numerical')
plt.plot(x,dy,label='exact')
```

Spot on...

## Unconstrained maximization 

Take the same function. We are looking for the maximum. Of course it is given by :
    $$ f'(x) = \alpha - \beta x^* = 0 $$
so $x^* = \frac{\alpha}{\beta}$. 

```python
xopt = alpha/beta
xopt
```

We denote this the analytical solution. Of course, we can find it numerically. Let's do it over a grid of values: 

```python
dy = df(x,alpha,beta)
[(i,di) for i,di in zip(x,dy)]
```

We see that between 8.88 and 11.11, the derivate changes sign. We could search for the solution in that smaller interval:

```python
xx = np.linspace(8.88,11.11,10)
[(i,di) for i,di in zip(xx,df(xx,alpha,beta))]
```

Now we see that it is happening between 9.87 and 10.11. We can repeat until we get very close (the interval becomes very small). Another avenue is called bisection and a slightly more sophisticated version called brent's method. Bisection is simple (see [this article](https://fr.wikipedia.org/wiki/M%C3%A9thode_de_dichotomie) on wikipedia).

```python
op.bisect(df,np.min(x),np.max(x),args=(alpha,beta))
```

We can also maximize the function. Python has minimizers. We can minimize the negative of the function...

```python
fm = lambda x: -f(x,alpha,beta)
opt =op.minimize_scalar(fm,method='golden')
opt.x
```
